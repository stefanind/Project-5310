```python
import geopandas as gpd
from shapely.geometry import Point
import pandas as pd
```

Downloaded a ZIP code tabulation area GeoJSON from the US Census to get the zipcodes from using the latitude and longitude.

```python
zcta = gpd.read_file("tl_2020_us_zcta510.shp")
```

```python
df = pd.read_csv('apts_2019_clean.csv')
```

```python
df["geometry"] = gpd.points_from_xy(df['longitude'], df['latitude'])
gdf = gpd.GeoDataFrame(df, geometry="geometry", crs="EPSG:4326")
```

```python
zcta.columns
```

```python
zcta = zcta.to_crs(epsg=4326)
```

```python
result = gpd.sjoin(gdf, zcta[["ZCTA5CE10", "geometry"]], how="left", predicate="within")
```

```python
result = result.rename(columns={"ZCTA5CE10": "zipcode"})
```

```python
result.head()
```

```python
result = result.drop(columns=["index_right", "geometry"])
```

```python
result.head()
```

```python
result.info()
```

```python
apts = pd.DataFrame(result)
```

```python
apts.to_csv('apts_2019_clean.csv', index=False, encoding='utf8')
```
