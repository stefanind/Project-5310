Simply by observation, we can see that there are multiple addresses for some properties, meaning that they have multiple units available with varying bedroom amount. This is why there is a max price vs min price. It can clearly be seen that the min price is the relevant column that denotes the price for the amount of bedrooms, e.g., if it is a 1 bedroom unit and the building goes up to 3 bedroom units, the min price would be substantially lower than the max price because the max price represents the 3 bedroom unit. Therefore, when using price to compare against the apartment listings from 2019, we should be using min_price rather than max_price.  

I will remove max price and keep min price but change it to just 'price' to represent this.

```python
# drop max_price and change min_price to 'price'
apts = apts.drop(columns=['max_price'])
apts = apts.rename(columns={'min_price': 'price'})
```
